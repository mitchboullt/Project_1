```python
import scipy.stats as st
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
# path_imput = os.path.join("NBA_Finals_and_MVP.xlsx")
# df = pd.read_excel(path_imput)
draft_df = pd.read_excel('Resources/NBA Drafts.xlsx')

mvp = pd.read_excel('new_folder/NBA_Finals_and_MVP.xlsx')
```

```python
mvp_2 = mvp.dropna()
```

```python
avgheight = round(mvp_2['MVP Height (ft)'].mean(),2)
```

```python
mvp3 = mvp_2.sort_values('MVP Height (ft)')
bar_plot = mvp3.plot.bar(x = 'MVP Name', y = 'MVP Height (ft)', ylim = [5.75,7.5], figsize=(20, 15))
bar_plot.axhline(y=avgheight, color= 'green', linewidth=2,)
bar_plot.text(x= 0, y=avgheight + .05, s='average MVP height',size = 20)
bar_plot.set_ylabel('Height')
```

```python
plt.plot(y = 'MVP Height (ft)', x = 'Year')


```

```python
y = mvp3.loc[:,'MVP Height (ft)'].values.reshape(-1, 1)  
x = mvp3.loc[:,'Year'].values.reshape(-1, 1)  
linear_regressor = LinearRegression() 
linear_regressor.fit(x, y)  # perform linear regression
y_pred = linear_regressor.predict(x) 
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.title("MVP Heights over time")
plt.xlabel("Year")
plt.ylabel("Height")
plt.show()
```
